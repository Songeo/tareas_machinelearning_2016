```python
#otra forma de generar los datos
til = "Plot with real points"
n0 = 200
n1 = n0
n2 = n0
n3 = n0

mean0 = [2,2]
mean1 = [-2,2]
mean2 = [-2,-2]
mean3 = [2,-2]

ax = plt.gca()

cov1 = [[1, 0], [0, 1]]  # diagonal covariance
x0, y0 = np.random.multivariate_normal(mean0, cov1, n0).T
x1, y1 = np.random.multivariate_normal(mean1, cov1, n1).T
x2, y2 = np.random.multivariate_normal(mean2, cov1, n2).T
x3, y3 = np.random.multivariate_normal(mean3, cov1, n3).T
plt.plot(x0, y0, '*b')
plt.plot(x1, y1, 'sr')
plt.plot(x2, y2, 'xg')
plt.plot(x3, y3, 'om')
plt.title(til)
plt.axis([-4, 4, -4, 4])
#plt.axis('equal')
#plt.show()

### Plot the circles
#circle 0:
c0 = mean0
r0 = 1
circle0 = plt.Circle((c0[0],c0[1]), r0, color='b', fill=False)
#circle 1:
c1 = mean1
r1 = 1
circle1 = plt.Circle((c1[0],c1[1]), r1, color='r', fill=False)
#circle 2:
c2 = mean2
r2 = 1
circle2 = plt.Circle((c2[0],c2[1]), r2, color='g', fill=False)
#circle 3:
c3 =  mean3
r3 = 1
circle3 = plt.Circle((c3[0],c3[1]), r3, color='m', fill=False)

ax.add_artist(circle0)
ax.add_artist(circle1)
ax.add_artist(circle2)
ax.add_artist(circle3)
```

```python
#print type(x0)
xx1 = np.concatenate((x0, x1, x2, x3), axis=0)
xx2 = np.concatenate((y0, y1, y2, y3), axis=0)

ax = plt.gca() #to plot the circles later on...
circle0 = plt.Circle((c0[0],c0[1]), r0, color='b', fill=False)
circle1 = plt.Circle((c1[0],c1[1]), r1, color='r', fill=False)
circle2 = plt.Circle((c2[0],c2[1]), r2, color='g', fill=False)
circle3 = plt.Circle((c3[0],c3[1]), r3, color='m', fill=False)

xx1_0c = xx1-c0[0]
xx2_0c = xx2-c0[1]
# print xx1_0c.size
# print xx2_0c.size
ratios0 = np.power(xx1_0c, 2) + np.power(xx2_0c, 2)
ratios1 = np.power(xx1-c1[0], 2) + np.power(xx2-c1[1], 2)
ratios2 = np.power(xx1-c2[0], 2) + np.power(xx2-c2[1], 2)
ratios3 = np.power(xx1-c3[0], 2) + np.power(xx2-c3[1], 2)
#y =  ( ratios0 <= (r0**2) ) or ( ratios1 <= (r1**2) )
yr0 = ratios0 <= (r0**2)
yr1 = ratios1 <= (r1**2)
yr2 = ratios2 <= (r2**2)
yr3 = ratios3 <= (r3**2)

y = np.logical_or(yr0, yr1)
y = np.logical_or(y, yr2)
y = np.logical_or(y, yr3)
#print y

xx1_insideCircle = xx1[y]
xx2_insideCircle = xx2[y]
#xx1_insideCircle = xx1[np.where(y==1)]
#xx2_insideCircle = xx2[np.where(y==1)]
# print '-----------------------'
# print 'xx1_insideCircle: '
# print xx1_insideCircle
# print '-----------------------'
# print 'x0: '
# print x0
# print '-----------------------'

#xx1_outsideCircle = xx1[np.where(y==0)]
#xx2_outsideCircle = xx2[np.where(y==0)]
xx1_outsideCircle = xx1[~y]
xx2_outsideCircle = xx2[~y]

#plot the inside-outsideCircle Points:
plt.plot(xx1_insideCircle,xx2_insideCircle,'*b')
plt.plot(xx1_outsideCircle,xx2_outsideCircle,'or')
plt.title('Points inside-outside the circle')
#and add the circle
ax.add_artist(circle0)
ax.add_artist(circle1)
ax.add_artist(circle2)
ax.add_artist(circle3)
```

```python
X = np.column_stack((xx1,xx2))
#print X 
#plt.plot(X[:,0],X[:,1],'sb')  #just to check things are working fine!
```

```python
#We predict - Kneighbours, SVM, and Neural Network
neigh = KNeighborsRegressor(n_neighbors=1)  #2
neigh.fit(X, y)

#Ojo, vamos a testear con el que entrenamos, no la mejor práctica.
predictNeigh = neigh.predict(X)

# print y
# print predictNeigh
print 'confusionmatrix = ', confusion_matrix(y, predictNeigh)
print 'precision = ', mtrcs.precision_score(y, predictNeigh)
print 'recall = ', mtrcs.recall_score(y, predictNeigh)
print 'Accuracy Score = ', mtrcs.accuracy_score(y, predictNeigh)

#la predicción es perfecta sobre el Training set, buena señal, aunque mala praxis
```

```python
#We predict - Kneighbours using X_train test etc...
neighDivide = KNeighborsRegressor(n_neighbors=1)  #2

X_train, X_test,Y_train,Y_test = train_test_split(X,y, train_size=0.75)
neighDivide.fit(X_train, Y_train)

#Ojo, vamos a testear con el que entrenamos, no la mejor práctica.
predictNeigh_test = neighDivide.predict(X_test)

# print y
# print predictNeigh
print 'confusionmatrix = ', confusion_matrix(Y_test, predictNeigh_test)
print 'precision = ', mtrcs.precision_score(Y_test, predictNeigh_test)
print 'recall = ', mtrcs.recall_score(Y_test, predictNeigh_test)
print 'Accuracy Score = ', mtrcs.accuracy_score(Y_test, predictNeigh_test)

X_hat_inside = X_test[predictNeigh_test==1,:]
X_hat_outside = X_test[predictNeigh_test==0,:]

#plot de predicciones:
ax = plt.gca() #to plot the circles later on...
circle0 = plt.Circle((c0[0],c0[1]), r0, color='b', fill=False)
circle1 = plt.Circle((c1[0],c1[1]), r1, color='r', fill=False)
circle2 = plt.Circle((c2[0],c2[1]), r2, color='g', fill=False)
circle3 = plt.Circle((c3[0],c3[1]), r3, color='m', fill=False)
#plot the inside-outsideCircle Points:
plt.title('Predicciones hechas con el K-neighbours algorithm :)')
plt.plot(X_hat_inside[:,0],X_hat_inside[:,1],'*b')
plt.plot(X_hat_outside[:,0],X_hat_outside[:,1],'or')
#and add the circle
ax.add_artist(circle0)
ax.add_artist(circle1)
ax.add_artist(circle2)
ax.add_artist(circle3)
plt.axis([-6, 6, -6, 6])
```

```python
### Aquí hacemos la predicción de SVM:
penalty = 0.3  #esta variable determina qué tanto "margen dejar"
clf = svm.SVC(kernel="rbf",probability=True, C=penalty)
clf.fit(X_train, Y_train)
# "probabilities"
probas_X_train = clf.predict_proba(X_train)
probas_X_test = clf.predict_proba(X_test)
#probas_X_test[1:10]

y_predSVM = clf.predict(X_test)
confusion_matrix(Y_test, y_predSVM)

#print clf.intercept_[0]
#print clf.coef_
print 'El cutoff fue de = ', cutoff, 'pero se puede hacer más estricto y subir el recall'
print 'confusionmatrix = ', confusion_matrix(Y_test, y_predSVM)
print 'precision = ', mtrcs.precision_score(Y_test, y_predSVM)
print 'recall = ', mtrcs.recall_score(Y_test, y_predSVM)
print 'Accuracy Score = ', mtrcs.accuracy_score(Y_test, y_predSVM)

cutoff = 0.5
X_hat_inside = X_test[y_predSVM>cutoff,:]
X_hat_outside = X_test[y_predSVM<cutoff,:]

#plot de predicciones:
ax = plt.gca() #to plot the circles later on...
circle0 = plt.Circle((c0[0],c0[1]), r0, color='b', fill=False)
circle1 = plt.Circle((c1[0],c1[1]), r1, color='r', fill=False)
circle2 = plt.Circle((c2[0],c2[1]), r2, color='g', fill=False)
circle3 = plt.Circle((c3[0],c3[1]), r3, color='m', fill=False)
#plot the inside-outsideCircle Points:
plt.title('Predicciones hechas con el SVM :)')
plt.plot(X_hat_inside[:,0],X_hat_inside[:,1],'*b')
plt.plot(X_hat_outside[:,0],X_hat_outside[:,1],'or')
#and add the circle
ax.add_artist(circle0)
ax.add_artist(circle1)
ax.add_artist(circle2)
ax.add_artist(circle3)
plt.axis([-6, 6, -6, 6])
```
