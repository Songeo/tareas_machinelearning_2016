```python
from sklearn.cross_validation import train_test_split
from sklearn import preprocessing, svm
from sklearn.metrics import confusion_matrix
from random import random, randint, seed, sample
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pda
import numpy as np
```

```python
dfand = pda.read_csv("../datos_clase/andSVM.csv")
dfand
```

```python
X = np.asarray( dfand[dfand.columns[:2]] )
y = np.asarray( dfand[dfand.columns[2]] )
svm_clf = svm.SVC(kernel='linear', probability=True, C = 9)
svm_clf.fit(X, y)  
```

```python
svm_clf.kernel
```

```python
wslist = [svm_clf.coef0, svm_clf.coef_[0][0], svm_clf.coef_[0][1]]
wslist
```

```python
fit_and = svm_clf.predict(X)
confusion_matrix(y, fit_and)
```

```python
svm_clf.support_vectors_
```

```python
ws0 = svm_clf.intercept_[0]
ws = svm_clf.coef_[0]
X1s = np.linspace(-2, 4, num=100)
X2s = (-ws[0]*X1s - ws0 ) / float(ws[1])

a = -ws[0] / ws[1]
b_down = svm_clf.support_vectors_[0]
y_down = a * X1s + (b_down[1] - a * b_down[0])

b_up = svm_clf.support_vectors_[-1]
y_up = a * X1s + (b_up[1] - a * b_up[0])
```

```python
plt.scatter(dfand[dfand.columns[0]], 
            dfand[dfand.columns[1]], 
            c = fit_and, s =60)
plt.plot(X1s ,X2s, '-k')
plt.plot(X1s, yy_down, 'k--')
plt.plot(X1s, yy_up, 'k--')
x2 = np.linspace(-1,2,100) # 100 numeros espaciados
#plt.plot(x2,-ws[0]/ws[2]+m*x2,color='red')
plt.xlabel("Obs")
plt.ylabel("Fit")
```

### CÃ­rculo

```python
dfcirc = pda.read_csv("../datos_clase/circulo.csv")
dfcirc.head()
```

```python
x1 = dfcirc[dfcirc.columns[:2]]#**2
y1 = dfcirc[dfcirc.columns[2]]
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, train_size=0.75)
```

```python
svm_circ = svm.SVC(kernel="rbf", probability=True, C=.3)
svm_circ.fit(x1_train, y1_train)  
fit_train = svm_circ.predict(x1_train)
fit_test = svm_circ.predict(x1_test)
```

```python
probs_train = svm_circ.predict_proba(x1_train)
probs_test = svm_circ.predict_proba(x1_test)
probs_train[1:5]
```

```python
confusion_matrix(y1_train, fit_train)
```

```python
plt.scatter(x1_train['X1'], x1_train['X2'], c = fit_train, 
            edgecolor = 'none',
            alpha = 0.5)
plt.colorbar(label = 'probabilidad')
```

```python
confusion_matrix(y1_test, fit_test)
```

```python
plt.scatter(x1_test['X1'], x1_test['X2'], c = fit_test, 
            edgecolor = 'none',
            alpha = 0.5)
plt.colorbar(label = 'probabilidad')
```
